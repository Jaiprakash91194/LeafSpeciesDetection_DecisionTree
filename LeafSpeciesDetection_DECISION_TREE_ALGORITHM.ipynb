```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
#Load dataset 
X, Y = load_iris(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state = 10)
```

```python
x_train.shape
```

#Since Decision tree is not affected by Scaling, not performing Feature Scaling.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
accuracy = []
for i in range(1,11):
  model = DecisionTreeClassifier(max_depth=i, random_state=3)
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accurancy_value = accuracy_score(y_pred, y_test)
  accuracy.append(accurancy_value)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6),dpi=75)
plt.plot(range(1,11), accuracy,"g-.v", mfc="r", ms=10)
plt.xlabel("MaxDepth value")
plt.ylabel("Accuracy Score")
plt.title("MaxDepth Vs AccuracyScore")
plt.grid()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model = DecisionTreeClassifier(max_depth=4, random_state=15)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
stack_array = np.column_stack((y_test, y_pred))
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(12,5))
plt.bar(range(1,11), accuracy, color=list("rgbycgrbyc"), width=0.6)
plt.ylim(0.900, 0.999)
plt.title("Model Performance")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy Rate")
plt.show()
```
